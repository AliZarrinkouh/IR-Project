```python
doc=data['title'].values # data is your dataframe
# tf-idf
vectorizer = TfidfVectorizer()
tfidf_doc = vectorizer.fit_transform(doc)
```

```python
query = 'MLS'
tfidf_query = vectorizer.transform([query])[0]
```

```python
type(tfidf_query), tfidf_query.shape
```

```python
# similarities
cosines = []
for q in tqdm(tfidf_doc):
    cosines.append(float(cosine_similarity(q, tfidf_query)))
```

```python
# sorting
k = 10
sorted_ids = np.argsort(cosines)
for i in range(k):
    cur_id = sorted_ids[-i-1]
    print(doc[cur_id], cosines[cur_id]) 
```
